####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
def revenue_cal(row):
    if row ['discounted_price'] > 600:
        return row['discounted_price'] * 0.25
    elif row['discounted_price'] > 350 and row['discounted_price'] <= 600:
        return 0.15 * row['discounted_price']
    elif row['discounted_price'] > 100 and row['discounted_price'] <= 350:
        return 0.10 * row['discounted_price']
    else:
        return 0.05 * row['discounted_price']
```

```python
ECom_Data['Revenue'] = ECom_Data.apply(lambda row: revenue_cal(row), axis=1)
```

```python
Total_Revenue = ECom_Data ['Revenue']. sum ()
print( 'Total Revenue', Total_Revenue)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days

### Ans 9 a)

```python
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'])
ECom_Data['month'] = ECom_Data['Order_Date'].dt.month
```

```python

plt.figure(figsize=(20,10))
a=sns.lineplot(ECom_Data['month'],ECom_Data['Revenue'],hue = ECom_Data['Region'],ci=False)
a.set_xlabel('Month',fontsize = 20)
a.set_ylabel('Revenue',fontsize = 20)
plt.title("Monthly Revenue by Region",fontsize = 20)
plt.legend()
plt.xticks(np.arange(1,13),["Jan","Feb","Mar","Apr","May","June","July","Aug","Sept","Oct","Nov","Dec"],fontsize = 15);
plt.show()
```

### Ans 9 b)

```python
ECom_Data['Order_Month']=pd.to_datetime(ECom_Data['Order_Date'], dayfirst=True).dt.to_period('M')
ECom_Data['Order_Month']=ECom_Data['Order_Month'].astype(str)

ECom_Monthly = pd.DataFrame(ECom_Data.groupby(['Order_Month','Region'])['Revenue'].sum().reset_index().sort_values(by='Order_Month'))

for reg in ECom_Monthly['Region'].unique():
    bmonth=(ECom_Monthly[ECom_Monthly['Region']==reg].sort_values(by='Revenue', ascending=False)[:1]['Order_Month'].values[0])
    wmonth=(ECom_Monthly[ECom_Monthly['Region']==reg].sort_values(by='Revenue', ascending=True)[:1]['Order_Month'].values[0])
    print('best  month for Region-{} is {}'.format(reg,bmonth))
    print('worst month for Region-{} is {}'.format(reg,wmonth))
```
